```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, balanced_accuracy_score, matthews_corrcoef, roc_auc_score
import pandas as pd
#Mean Absolute Difference=Fisher's Score
# Define your dataset and target column here
target_column = 'cardio'

# Define features (X) and target variable (y)
X = cdata.drop(columns=[target_column])
y = cdata[target_column]

# Selected features from different feature selection methods
selected_features_chi_square = ['age(in years)', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'bmi', 'bmi_category', 'pulse_pressure', 'map']
selected_features_info_gain = ['age(in years)', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'bmi', 'bmi_category', 'pulse_pressure', 'map']
selected_features_fishers_score = ['age(in years)', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'bmi', 'bmi_category', 'pulse_pressure', 'map']
selected_features_variance_threshold = ['age(in years)', 'gender', 'height', 'weight', 'ap_hi', 'smoke','alco','ap_lo', 'cholesterol', 'gluc', 'active', 'bmi', 'bmi_category', 'pulse_pressure', 'map']

selected_feature_sets = [
    ('Chi-Square Test', selected_features_chi_square),
    ('Information Gain', selected_features_info_gain),
    ("Fisher's Score", selected_features_fishers_score),
    ('Variance Threshold', selected_features_variance_threshold)
]

# Initialize different boosting classifiers
classifiers = [
    ('Gradient Boosting', GradientBoostingClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('XGBoost', XGBClassifier()),
    ('LightGBM', LGBMClassifier()),
    ('CatBoost', CatBoostClassifier())
]

# Initialize an empty list to store results
results = []

# Iterate through each feature set and each classifier
for feature_method, features in selected_feature_sets:
    for classifier_name, classifier in classifiers:
        print(f"Feature Selection Method: {feature_method}, Classifier: {classifier_name}")

        X_selected = X[features]  # Select features based on the method

        # Split the dataset into train and test sets using selected features
        X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

        # Train the classifier
        classifier.fit(X_train, y_train)

        # Make predictions
        y_pred = classifier.predict(X_test)

        # Calculate evaluation metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
        matthews_corr = matthews_corrcoef(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_pred)

        # Store results
        result_entry = {
            'Feature Selection Method': feature_method,
            'Classifier': classifier_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
            'Balanced Accuracy': balanced_accuracy,
            'Matthews Corrcoef': matthews_corr,
            'ROC AUC': roc_auc
        }
        results.append(result_entry)

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Save results to a CSV file
results_df.to_csv('boosting_algorithm_results_with_feature_selection.csv', index=False)
```
